# Insights and Recommendations

This notebook summarizes the key findings from the Global Cancer Analytics project and translates statistical analysis into actionable insights and recommendations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/raw/global_cancer_patients.csv")

data.head()
```

## Project Objective

The objective of this project was to analyze global cancer patient data to understand:

- Factors influencing cancer severity
- Patterns in early-stage diagnosis
- Economic burden across demographics and countries
- Relationship between treatment cost and survival outcomes

The analysis used statistical and inferential methods to derive meaningful healthcare insights.

## Key Insights

### 1. Risk Factors and Cancer Severity
Lifestyle and genetic risk factors show stronger associations with cancer severity compared to demographic variables.

### 2. Early Diagnosis Patterns
Approximately 40% of cancer cases are diagnosed at an early stage across cancer types, indicating moderate effectiveness of early detection systems.

### 3. Economic Burden
Treatment costs vary significantly across countries and increase with patient age, highlighting healthcare inequality and financial challenges.

### 4. Treatment Cost vs Survival
Statistical hypothesis testing showed no significant relationship between treatment cost and survival years, suggesting that higher spending alone does not guarantee better outcomes.

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    x="Target_Severity_Score",
    y="Treatment_Cost_USD",
    data=data,
    alpha=0.3
)

plt.title("Severity vs Treatment Cost Overview")
plt.show()
```

## Major Conclusions

- Cancer outcomes are influenced by multiple interacting factors rather than a single dominant predictor.
- Early detection plays a critical role in reducing both medical severity and financial burden.
- Economic disparities across countries significantly impact treatment costs.
- Healthcare effectiveness appears more important than treatment expenditure alone.

## Recommendations

Based on the analysis, the following recommendations are suggested:

- Strengthen early cancer screening and awareness programs.
- Focus preventive strategies on lifestyle-related risk factors.
- Improve healthcare accessibility in regions with lower early diagnosis rates.
- Provide financial support programs for elderly cancer patients.
- Encourage investment in efficient treatment methods rather than only increasing spending.

## Limitations

- The dataset represents observational data; correlation does not imply causation.
- Some clinical variables that may influence survival were not available.
- Results depend on dataset assumptions and may vary with real-world clinical data.
